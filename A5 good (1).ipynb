```python
import pandas as pd
import networkx as nx
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules,apriori
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"D:\ACADEMICS\TE-IT SEM5\ml lab\5.Market_Basket_Optimisation.csv",header=None)
```

```python
data
```

```python
transactions=[]
for i in range(len(data)):
    transaction=[]
    for j in range(len(data.columns)):
        if str(data.values[i,j])!='nan':
            transaction.append(str(data.values[i,j]))
    transactions.append(transaction)        
```

```python
te=TransactionEncoder()
x=te.fit_transform(transactions)
```

```python
df=pd.DataFrame(x,columns=te.columns_)
```

```python
df
```

```python
frequent_item_set=apriori(df,min_support=0.01,use_colnames=True)
```

```python
rules=association_rules(frequent_item_set,min_threshold=0.10)
```

```python
rules.head(100)
```

```python
rules[rules['antecedents']=={'cake'}]
```

```python
plt.scatter(rules['confidence'],rules['support'],alpha=0.5)
```

```python
G = nx.DiGraph()
plt.figure(figsize=(20,20))
for index, row in rules.iterrows():
    antecedent = list(row['antecedents'])[0]
    consequent = list(row['consequents'])[0]
    confidence = row['confidence']
    
    G.add_edge(antecedent,consequent,confidence=confidence)

pos=nx.spring_layout(G,k=15)    
    

    
nx.draw(G,pos,with_labels=True, node_size=1000, node_color='lightblue', font_size=12, font_color='black')
nx.draw_networkx_edge_labels(G, pos, font_size=8, font_color='red')
plt.show()
```
