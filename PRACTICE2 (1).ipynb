```python
x=df.loc[:,"YEAR"].values.reshape(-1,1)
y=df.loc[:,"ANNUAL"].values.reshape(-1,1)
plt.plot(x,y,"*")
```

```python
type(x)
type(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_test_prediction=regressor.predict(x_test)
y_train_prediction=regressor.predict(x_train)
```

```python
#model visualize
plt.scatter(x_test,y_test,c="red")
plt.xlabel("YEAR")
plt.ylabel("TEMPERATURE")
#regression line
plt.plot(x_test,y_test_prediction,"-")
```

```python
#test prediction error
MAE=mean_absolute_error(y_test,y_test_prediction)
print("MAE",MAE)
MSE = mean_squared_error(y_test,y_test_prediction)
print("MSE",MSE)
r1 = r2_score(y_test,y_test_prediction)
print("RSquare" , r1)

#train prediction error
MAE=mean_absolute_error(y_train,y_train_prediction)
print("MAE",MAE)
MSE = mean_squared_error(y_train,y_train_prediction)
print("MSE",MSE)
r1 = r2_score(y_train,y_train_prediction)
print("RSquare" , r1)
```

## COMPARISON OF DIFFERENT TEST SIZE MODELS

```python
train_sse=[]
test_sse=[]
for i in range (1,10):
    i/=10.0
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=i,random_state=2)
    model=LinearRegression()
    model.fit(x_train,y_train)
    
    y_test_predicted=model.predict(x_test)
    y_train_predicted=model.predict(x_train)
    
    train_sse.append(np.sum((y_train_predicted-y_train)**2)) 
    test_sse.append(np.sum((y_test_predicted-y_test)**2))
plt.plot(test_sse,'b',label="test")
plt.plot(train_sse,'r',label="train")
plt.show()
```
